```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
sT = data[" Settle Price"]
sT = np.array(sT)
sT = np.reshape(sT, (18,))
print(sT)
```

```python
# stock price
spot_price = 1490.5
# Long future
strike_price_long_future = 1491.50
premium_long_future = 0
# Stock price range at expiration of the put
```

```python
def call_payoff(sT, strike_price, premium):
    return np.where(sT > strike_price, sT - strike_price, sT - strike_price) - premium

payoff_long_future = call_payoff (sT, strike_price_long_future, premium_long_future)
# Plot
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False) # Top border removed
ax.spines['right'].set_visible(False) # Right border removed
ax.spines['bottom'].set_position('zero') # Sets the X-axis in the center
ax.plot(sT,payoff_long_future,label='Long Call',color='g')
plt.xlabel('Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
def short_payoff(sT, strike_price, premium):
    return np.where(strike_price > sT, strike_price - sT, strike_price - sT) - premium

payoff_long_future = short_payoff (sT, strike_price_long_future, premium_long_future)
# Plot
fig, ax = plt.subplots()
ax.spines['top'].set_visible(False) # Top border removed
ax.spines['right'].set_visible(False) # Right border removed
ax.spines['bottom'].set_position('zero') # Sets the X-axis in the center
ax.plot(sT,payoff_long_future,label='Short Call',color='g')
plt.xlabel('Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```
